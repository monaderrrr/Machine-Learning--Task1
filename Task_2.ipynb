```python
# Mohamed Nader Rashad
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[4], [5], [7], [1], [3]])
y = np.array([60, 70, 80, 85, 90])

model = LinearRegression()
model.fit(X, y)

X_new = np.array([[6]])
y_pred = model.predict(X_new)
print(f"Predicted grade for 6 hours: {y_pred[0]:.2f}")

plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, model.predict(X), color='red', label='Fit Line')
plt.xlabel('Study Hours')
plt.ylabel('Grades')
plt.title('Linear Regression')
plt.legend()
plt.show()
# Mohamed Nader Rashad
```

```python
# Mohamed Nader Rashad
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[90], [10], [50], [110], [130]])
y = np.array([0, 0, 1, 1, 1])

model = LogisticRegression()
model.fit(X, y)

X_new = np.array([[100]])
y_pred = model.predict(X_new)
print(f"Prediction for 100 sqm: {'Sold' if y_pred[0] == 1 else 'Not Sold'}")

X_plot = np.linspace(40, 140, 100).reshape(-1, 1)
y_prob = model.predict_proba(X_plot)[:, 1]
plt.scatter(X, y, color='green', label='Data')
plt.plot(X_plot, y_prob, color='orange', label='Sale Probability')
plt.xlabel('House Size (sqm)')
plt.ylabel('Sale Probability')
plt.title('Logistic Regression')
plt.legend()
plt.show()
# Mohamed Nader Rashad
```
